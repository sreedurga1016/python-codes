```python
#+ve and -ve numbers
num = float(input("enter a number :"))
if num>0:
  print("positive")
elif num == 0:
  print("zero")
else:
  print("negative")
```

```python
#even numbers
num=int(input("enter a number:"))
if num%2==0:
  print("even")
else:
  print("odd")
```
